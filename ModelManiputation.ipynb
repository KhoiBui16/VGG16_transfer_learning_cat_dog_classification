## Lấy ra featuers của VGG16 và tạo ra new Classifier ?

---
**Ưu điểm**:
- Phù hợp với bài toán của mình riêng phù hợp với bài toán của mình từ 1 features có sẵn và tão ra classifier mới

```python
import torch.nn as nn
import torchvision.models as models
from torchvision.models import VGG16_Weights

# vgg16 = models.vgg16(pretrained=True) # Đây là cách phiên bản cũ

# Nên dùng cách phiên bản mới
vgg16 = models.vgg16(weights=VGG16_Weights.IMAGENET1K_V1)
f_extractor = vgg16.features

# Create new model
model = nn.Sequential(
    f_extractor,
    nn.Flatten(),
    nn.Linear(512 * 1 * 1, 10))

model = model.to(device)

summary(model, (3, 32, 32))
```

**Lý do lấy `pretrained=True`**:
- 

- Lấy trọng số của mô hình đã train rồi -> accuracy đã tốt và đây là được train trên bộ dataset lớn -> tốt và tớn thời gian
- Đây gọi là bài toán `Transfer Learning` -> lấy ra **trọng số có sẵn**
