# Load Dataset

```python
import os
import numpy as np
import cv2
from random import shuffle


folder1 = 'data/train/cats'
IMG_SIZE = 50

# training data
X_train = []
y_train = []


# listdir -> return all file' names
for img_name in os.listdir(folder1):
    label = 0
    # create path to each file
    path = os.path.join(folder1, img_name)

    # Read image
    img = cv2.imread(path, 1)
    assert img is not None, f"❌ Không đọc được ảnh: {path}"
    
    # convert BGR to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
    
    # Resize img
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    
    # save image
    X_train.append(img)
    y_train.append(label)
    
X_train = np.array(X_train)
y_train = np.array(y_train)

"""
    - Lưu data theo numpy thay vì rời rạc mặc dù tổng bộ nhớ của cách lưu bộ nhớ sẽ lớn hơn so với cách cũ:
        + Load nhanh hơn vì đây đã lưu ở cache -> ngay kế cpu -> tốc độ nhanh hơn
        + còn cách load cũ -> phải đọc file ngoài -> xa cpu -> lâu hơn
"""

if not os.path.exists('data/X_train.npy') or not os.path.exists('data/y_train.npy'):
    np.save('data/X_train.npy', X_train)
    np.save('data/y_train.npy', y_train)

print(X_train.shape)
print(y_train.shape)
```

```python
# convert to torch
import torch

X_train = torch.from_numpy(X_train)
y_train = torch.from_numpy(y_train)

print(X_train.shape)
print(y_train.shape)
```

# Cách đọc cũng nhiều folder

```python
import os
import numpy as np
import cv2
from random import shuffle
from tqdm import tqdm


folder1 = 'data/train/cats'
IMG_SIZE = 50

# training data
X_train = []
y_train = []


# listdir -> return all file' names
for img_name in tqdm(os.listdir(folder1)):
    label = 0
    # create path to each file
    path = os.path.join(folder1, img_name)

    # Read image
    img = cv2.imread(path, 1)
    assert img is not None, f"❌ Không đọc được ảnh: {path}"
    
    # convert BGR to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
    
    # Resize img
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    
    # save image
    X_train.append(img)
    y_train.append(label)
    
X_train = np.array(X_train)
y_train = np.array(y_train)

"""
    - Lưu data theo numpy thay vì rời rạc mặc dù tổng bộ nhớ của cách lưu bộ nhớ sẽ lớn hơn so với cách cũ:
        + Load nhanh hơn vì đây đã lưu ở cache -> ngay kế cpu -> tốc độ nhanh hơn
        + còn cách load cũ -> phải đọc file ngoài -> xa cpu -> lâu hơn
"""

if not os.path.exists('data/X_train.npy') or not os.path.exists('data/y_train.npy'):
    np.save('data/X_train.npy', X_train)
    np.save('data/y_train.npy', y_train)

print(X_train.shape)
print(y_train.shape)
```
