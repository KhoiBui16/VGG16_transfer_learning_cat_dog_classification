# Load Dataset

```python
import os
import numpy as np
import cv2
from random import shuffle


folder1 = 'data/train/cats'
IMG_SIZE = 50

# training data
X_train = []
y_train = []


# listdir -> return all file' names
for img_name in os.listdir(folder1):
    label = 0
    # create path to each file
    path = os.path.join(folder1, img_name)

    # Read image
    img = cv2.imread(path, 1)
    assert img is not None, f" Khng c c nh: {path}"
    
    # convert BGR to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
    
    # Resize img
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    
    # save image
    X_train.append(img)
    y_train.append(label)
    
X_train = np.array(X_train)
y_train = np.array(y_train)

"""
    - Lu data theo numpy thay v ri rc mc d tng b nh ca cch lu b nh s ln hn so vi cch c:
        + Load nhanh hn v y  lu  cache -> ngay k cpu -> tc  nhanh hn
        + cn cch load c -> phi c file ngoi -> xa cpu -> lu hn
"""

if not os.path.exists('data/X_train.npy') or not os.path.exists('data/y_train.npy'):
    np.save('data/X_train.npy', X_train)
    np.save('data/y_train.npy', y_train)

print(X_train.shape)
print(y_train.shape)
```

```python
# convert to torch
import torch

X_train = torch.from_numpy(X_train)
y_train = torch.from_numpy(y_train)

print(X_train.shape)
print(y_train.shape)
```

# Cch c cng nhiu folder

```python
import os
import numpy as np
import cv2
from random import shuffle
from tqdm import tqdm


folder1 = 'data/train/cats'
IMG_SIZE = 50

# training data
X_train = []
y_train = []


# listdir -> return all file' names
for img_name in tqdm(os.listdir(folder1)):
    label = 0
    # create path to each file
    path = os.path.join(folder1, img_name)

    # Read image
    img = cv2.imread(path, 1)
    assert img is not None, f" Khng c c nh: {path}"
    
    # convert BGR to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
    
    # Resize img
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    
    # save image
    X_train.append(img)
    y_train.append(label)
    
X_train = np.array(X_train)
y_train = np.array(y_train)

"""
    - Lu data theo numpy thay v ri rc mc d tng b nh ca cch lu b nh s ln hn so vi cch c:
        + Load nhanh hn v y  lu  cache -> ngay k cpu -> tc  nhanh hn
        + cn cch load c -> phi c file ngoi -> xa cpu -> lu hn
"""

if not os.path.exists('data/X_train.npy') or not os.path.exists('data/y_train.npy'):
    np.save('data/X_train.npy', X_train)
    np.save('data/y_train.npy', y_train)

print(X_train.shape)
print(y_train.shape)
```
